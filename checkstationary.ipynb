### **Preprocessing function for training data**

```python
def preprocess_data(data):
    # One-hot encoding for categorical variables
    categorical_columns = ['ProductType', 'Manufacturer', 'Area Code', 'Sourcing Channel', 'Product Type']
    data = pd.get_dummies(data, columns=categorical_columns, dtype=int)  # Specify dtype=int for integer encoding

    # Label encoding for 'Product Size' column
    label_encoder = LabelEncoder()
    data['Product Size'] = label_encoder.fit_transform(data['Product Size'])

    # Datetime conversion
    data['Month of Sourcing'] = pd.to_datetime(data['Month of Sourcing'], format='%b-%y')
    
    data['Sourcing Cost'] = data['Sourcing Cost'].astype(int)
    
    # Convert datetime to integer representation
    data['Month of Sourcing'] = data['Month of Sourcing'].apply(lambda x: x.timestamp()).astype(int)

    return data
```

### **This preprocessing function used in time of test**

```python

def preprocess_data_test(data):
    # One-hot encoding for categorical variables
    categorical_columns = ['ProductType', 'Manufacturer', 'Area Code', 'Sourcing Channel', 'Product Type']
    data = pd.get_dummies(data, columns=categorical_columns, dtype=int)

    # Label encoding for 'Product Size' column
    label_encoder = LabelEncoder()
    data['Product Size'] = label_encoder.fit_transform(data['Product Size'])
    # data['Month of Sourcing'] = label_encoder.fit_transform(data['Month of Sourcing'])

    # Datetime conversion
    data['Month of Sourcing'] = pd.to_datetime(data['Month of Sourcing'], format='%b-%y')
    
    # data['Sourcing Cost'] = data['Sourcing Cost'].astype(int)
    # Convert datetime to integer representation
    data['Month of Sourcing'] = data['Month of Sourcing'].apply(lambda x: x.timestamp())
    # Convert to integer
    data['Month of Sourcing'] = data['Month of Sourcing'].astype(int)

    return data

```

### **Spliting the trainig data in X which contain all the columns except sourcing cost and y contains only sourcing cost**

```python
X = train_data.drop(['Sourcing Cost'], axis=1)
y = train_data['Sourcing Cost']
```

### **Spliting the training data in which we have test_size=0.000001 which means training dataset used for pure training**

```python
X_train, _, y_train, _ = train_test_split(X, y, test_size=0.000001, random_state=42)
```

### **Spliting the testing data in X which contain all the columns except sourcing cost and y contains only sourcing cost**

```python
x = test_data.drop(['Sourcing Cost'], axis=1)
Y= test_data['Sourcing Cost']
```

### **Spliting the testing data in which we have test_size=0.98 which means testing dataset used for pure testing**

```python
_, x_test, _, Y_test = train_test_split(x, Y, test_size=0.98, random_state=42)
```

### **RandomForestRegressor**

```python
model = RandomForestRegressor(n_estimators=200, random_state=42)
model.fit(X_train, y_train)
```

### **Prediction and finding MSE value**

```python
y_pred = model.predict(x_test)
mse = mean_squared_error(Y_test, y_pred)
print("Mean Squared Error on Validation Set:", mse)
mae = mean_absolute_error(Y_test,y_pred)
print("Mean absolute Error on Validation Set:", mae)
```

### **LightGBM Model**

```python
# Convert data to LightGBM Dataset format
train_data = lgb.Dataset(X_train, label=y_train)
val_data = lgb.Dataset(x_test, label= Y_test, reference=train_data)

# Define LightGBM parameters
params = {
    'objective': 'regression',
    'metric': 'rmse',
    'num_leaves': 31,
    'learning_rate': 0.05,
    'early_stopping_rounds': 10,  # Specify early stopping rounds here
    'verbose': -1
}

# Train the LightGBM model
num_rounds = 200
model01 = lgb.train(params, train_data, num_rounds, valid_sets=[val_data])

# Predict on validation set
y_pred01 = model01.predict(x_test, num_iteration=model01.best_iteration)

```

```python
mse = mean_squared_error(Y_test, y_pred01)
print("Mean Squared Error on Validation Set:", mse)
mae = mean_absolute_error(Y_test,y_pred01)
print("Mean absolute Error on Validation Set:", mae)
```

### **XGBoost Model**

```python
# Convert data to XGBoost DMatrix format
dtrain = xgb.DMatrix(X_train, label=y_train)
dval = xgb.DMatrix(x_test, label=Y_test)

# Define XGBoost parameters
params = {
    'objective': 'reg:squarederror',  # for regression tasks
    'eval_metric': 'rmse',             # Root Mean Squared Error
    'eta': 0.05,                       # learning rate
    'max_depth': 6,                    # maximum depth of the tree
    'min_child_weight': 1,             # minimum sum of instance weight (hessian) needed in a child
    'early_stopping_rounds': 10,       # specify the number of rounds to wait for early stopping
    'verbose': False                   # verbosity
}

# Train the XGBoost model
num_rounds = 200
model02 = xgb.train(params, dtrain, num_rounds, evals=[(dtrain, 'train'), (dval, 'validation')], early_stopping_rounds=params['early_stopping_rounds'])

# Predict on validation set
y_pred02 = model02.predict(dval)

```

### **Finding MSE & MAE**

```python
mse = mean_squared_error(Y_test, y_pred02)
print("Mean Squared Error on Validation Set:", mse)
mae = mean_absolute_error(Y_test,y_pred02)
print("Mean absolute Error on Validation Set:", mae)
```
